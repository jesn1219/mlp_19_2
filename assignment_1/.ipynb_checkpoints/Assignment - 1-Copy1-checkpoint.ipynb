```python

train_pool_size = 500 # Actually, datasets will generate overall 1000. ( 500 * A, 500 * B)
test_pool_size = 500
```

```python
class point :
    def __init__(self,x,y) :
        self.set_x(x)
        self.set_y(y)
        print("Point is setted. X : %d | Y : %d"%(self.x,self.y))
    def set_x(self,x) :
        self.x = x
    def set_y(self,y) :
        self.y = y
    def get_x(self) :
        return self.x
    def get_y(self) :
        return self.y
```

```python
class unit :
    def __init__(self, x, y, label) :
        self.x = x
        self.y = y
        self.label = label
        #print("test_unit is created. X : %d | Y : %d"%(self.x,self.y))
     
    def set_x(self,x) :
        self.x = x
    def set_y(self,y) :
        self.y = y
    def get_x(self) :
        return self.x
    def get_y(self) :
        return self.y    
    def get_label(self) :
        return self.label
```

```python

# centroid setting

train_point_A = point(25,40)
train_point_B = point(35,30)

train_pool_a = []
train_pool_b = []
test_pool_a = []
test_pool_b = []
np.random.seed(100)

# Random datasets generate

for i in range(0,train_pool_size) :
    random_x = train_point_A.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_A.get_y() + (5 * ((np.random.randn(1))[0]))
    train_pool_a.append(unit(random_x,random_y,0.0))
    random_x = train_point_B.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_B.get_y() + (5 * ((np.random.randn(1))[0]))
    train_pool_b.append(unit(random_x,random_y,1))
    
train_pool = train_pool_a+train_pool_b    

for i in range(0, test_pool_size) :
    random_x = train_point_A.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_A.get_y() + (5 * ((np.random.randn(1))[0]))
    test_pool_a.append(unit(random_x,random_y,0))
    random_x = train_point_B.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_B.get_y() + (5 * ((np.random.randn(1))[0]))
    test_pool_b.append(unit(random_x,random_y,1.0))
test_pool = test_pool_a + test_pool_b
```

```python
train_pool = train_pool_a + train_pool_b
test_pool = test_pool_a + test_pool_b

x_train = [u.get_x() for u in train_pool]  
y_train = [u.get_y() for u in train_pool]  
x_test = [u.get_x() for u in test_pool]
y_test = [u.get_y() for u in test_pool]
```

```python
t1 = plt.scatter(x_train,y_train, color='orange',marker='o')
t2 = plt.scatter(x_test, y_test, color= 'red', marker= 'o')
plt.title("Two Clusters of dataset")
plt.legend([t1,t2], ['Training dataset','testing dataset'])
plt.show()
```

```python
# sigmoid Function.
def sigmoid(z) :
    return 1 / (1 + math.e ** (-z))
```

```python
# h(x) Function. hypothesis Func.
def hypothesis(weight,X) :
    theta = np.array(weight)
    z = np.dot(X,theta)
    h = sigmoid(z)
    return h
```

```python
# Data reconstruct to make easy to calculate

X = []
y = []

# fetch type A units
for u in train_pool_a :
    X.append([u.get_x(),u.get_y(),0])
    y.append([0])
# type B units
for u in train_pool_b :
    X.append([u.get_x(),u.get_y(),1])
    y.append([1])
X = np.array(X)
y = np.array(y)


# Test datasets.. 
X_test = []
y_test = []
for u in test_pool_a :
    X_test.append([u.get_x(),u.get_y(),0])
    y_test.append([0])
for u in test_pool_b :
    X_test.append([u.get_x(),u.get_y(),1])
    y_test.append([1])
    
X_test = np.array(X_test)
y_test = np.array(y_test)

```

```python
# Loss Function
def lossFunction(h,y) :
    # if i = 1, it occurs ZeroDivisionError ,
    # So, adjust value to 0.995
    for i in h :
        if i == 1 :
            i = 0.995
    try : 
        ret = (-y * np.log(h) - (1-y) * np.log(1-h)).mean()
    except ZeroDivisionError :
        print("Error")
    finally :
        return (-y * np.log(h) - (1-y) * np.log(1-h)).mean()
```

```python
# predict Function. if h(x) returns >=0.5, set to 1. other cases, set to 0.
def predict(weight,pool) :
    m = len(pool)
    correct = 0
    for u in pool :
        x = u.get_x()
        y = u.get_y()
        l = u.label
        X = [[x,y,l]]
        predict = hypothesis(weight,X)
        if (predict >= 0.5) :
            predict = 1
        else :
            predict = 0
        
        if predict == u.label :
            correct += 1
    return correct * (1/m)
```

```python
# Initial Weight Value
weight = [[-0.001],[-0.001],[0]]

# Learning rate 
learning_rate = 0.00001

# Log variables setting
loss_log_train = []
loss_log_test = []
iter_log = []
acc_log_train = []
acc_log_test = []



m = train_pool_size
i = 0

# train, 5000 times
while (i < 5000):
    i += 1
    
    # get hypothesis values with entire X (train dataset)
    h = hypothesis(weight,X)
    
    # get hypothesis values with test dataset
    h_test = hypothesis(weight,X_test)
    
    # gradient calculate. 
    gradient = np.dot(X.T, h-y) / y.size
    
    # adjust weight values with gradient value
    weight-= learning_rate*gradient    
    
    # get train loss value
    cur_train_loss = lossFunction(h,y)
    # get test loss value
    test_loss = lossFunction(h_test,y_test)
    
    # get accuracy
    acc = predict(weight,train_pool)
    acc_test = predict(weight, test_pool)
    
    # add log data
    loss_log_train.append(cur_train_loss)
    loss_log_test.append(test_loss)
    iter_log.append(i)
    acc_log_train.append(acc)
    acc_log_test.append(acc_test)
    
    '''
    if (i%1000) :
        #print("loss: %f,%f , it: %d, t_acc : %f / t_acc : %f"%(cur_train_loss,test_loss,i,acc,acc_test)) 
    #print("weight : %f, %f, %f"%(weight[0],weight[1],weight[2]))
    '''
    
print("Finished")    
```

```python
t1 = plt.plot(iter_log,loss_log_train, color='orange',label='Training Loss')
t2 = plt.plot(iter_log,loss_log_test, color= 'red',label='Testing Loss')
plt.title("Loss")
plt.legend(['Training Loss','Testing Loss'])
plt.show()
```

```python
t1 = plt.plot(iter_log,acc_log_train, color='orange',label='Training Acc')
t2 = plt.plot(iter_log,acc_log_test, color= 'red',label='Testing Acc')
plt.title("Accuracy")
plt.legend(['Training Acc','Testing Acc'])
plt.show()
```
