```python
''' Hi , '''
import matplotlib.pyplot as plt
import math
import numpy as np
import random
```

```python
print("Hello")
train_pool_size = 500
test_pool_size = 500
```

```python
class point :
    def __init__(self,x,y) :
        self.set_x(x)
        self.set_y(y)
        print("Point is setted. X : %d | Y : %d"%(self.x,self.y))

    def set_x(self,x) :
        self.x = x
    def set_y(self,y) :
        self.y = y
    def get_x(self) :
        return self.x
    def get_y(self) :
        return self.y
```

```python
class unit :
    def __init__(self, x, y, label) :
        self.x = x
        self.y = y
        self.label = label
        #print("test_unit is created. X : %d | Y : %d"%(self.x,self.y))
     
    def set_x(self,x) :
        self.x = x
    def set_y(self,y) :
        self.y = y
    def get_x(self) :
        return self.x
    def get_y(self) :
        return self.y    
    def get_label(self) :
        return self.label
```

```python
train_point_A = point(15,30)
train_point_B = point(30,15)

train_pool_a = []
train_pool_b = []
test_pool_a = []
test_pool_b = []
np.random.seed(100)

for i in range(0,train_pool_size) :
    random_x = train_point_A.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_A.get_y() + (5 * ((np.random.randn(1))[0]))
    train_pool_a.append(unit(random_x,random_y,0.0))
    random_x = train_point_B.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_B.get_y() + (5 * ((np.random.randn(1))[0]))
    train_pool_b.append(unit(random_x,random_y,1))
    
train_pool = train_pool_a+train_pool_b    

for i in range(0, test_pool_size) :
    random_x = train_point_A.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_A.get_y() + (5 * ((np.random.randn(1))[0]))
    test_pool_a.append(unit(random_x,random_y,0))
    random_x = train_point_B.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_B.get_y() + (5 * ((np.random.randn(1))[0]))
    test_pool_b.append(unit(random_x,random_y,1.0))
test_pool = test_pool_a + test_pool_b
```

```python
train_pool = train_pool_a + train_pool_b
test_pool = test_pool_a + test_pool_b

x_train = [u.get_x() for u in train_pool]  
y_train = [u.get_y() for u in train_pool]  
x_test = [u.get_x() for u in test_pool]
y_test = [u.get_y() for u in test_pool]
```

```python
x_train_a = [u.get_x() for u in train_pool_a] 
y_train_a = [u.get_y() for u in train_pool_a]
x_train_b = [u.get_x() for u in train_pool_b]
y_train_b = [u.get_y() for u in train_pool_b]

t1 = plt.scatter(x_train_a, y_train_a, color='blue', marker='o')
t2 = plt.scatter(x_train_b, y_train_b, color='black', marker='o')
plt.title("Two Clusters of Training dataSet")
plt.legend([t1,t2],['A','B'])
plt.show()


```

```python
x_test_a = [u.get_x() for u in test_pool_a]
y_test_a = [u.get_y() for u in test_pool_a]
x_test_b = [u.get_x() for u in test_pool_b]
y_test_b = [u.get_y() for u in test_pool_b]

t1 = plt.scatter(x_test_a, y_test_a, color='blue', marker='o')
t2 = plt.scatter(x_test_b, y_test_b, color='black', marker='o')
plt.title("Two Clusters of Testing dataSet")
plt.legend([t1,t2],['A','B'])
plt.show()


```

```python
t1 = plt.scatter(x_train,y_train, color='orange',marker='o')
t2 = plt.scatter(x_test, y_test, color= 'red', marker= 'o')
plt.title("Two Clusters of dataset")
plt.legend([t1,t2], ['Training dataset','testing dataset'])
plt.show()
```

```python
def sigmoid(z) :
    return 1 / (1 + math.e ** (-z))
```

```python
def hypothesis(weight,X) :
    theta = np.array(weight[:2])
    b = weight[2]
    z = np.dot(X,theta) + b
    
    h = sigmoid(z)
    return h
```

```python
X = []
y = []
for u in train_pool_a :
    X.append([u.get_x(),u.get_y()])
    y.append([0.0])
for u in train_pool_b :
    X.append([u.get_x(),u.get_y()])
    y.append([1.0])
X = np.array(X)
y = np.array(y)

X_test = []
y_test = []

for u in test_pool_a :
    X_test.append([u.get_x(),u.get_y()])
    y_test.append([0.0])
for u in test_pool_b :
    X_test.append([u.get_x(),u.get_y()])
    y_test.append([1.0])

X_test = np.array(X_test)
y_test = np.array(y_test)


```

```python
def lossFunction(h,y) :
    for i in h :
        if i == 1 :
            i = 0.95
    
    try : 
        ret = (-y * np.log(h) - (1-y) * np.log(1-h)).mean()
    except ZeroDivisionError :
        print("Error")
    finally :
        return (-y * np.log(h) - (1-y) * np.log(1-h)).mean()
```

```python
def predict(weight,pool) :
    m = len(pool)
    correct = 0
    for u in pool :
        x = u.get_x()
        y = u.get_y()
        X = [[x,y]]
        predict = hypothesis(weight,X)
        #print("predict : %f, label : %d"%(predict,u.label))
        if (predict >= 0.5) :
            predict = 1
        else :
            predict = 0
        
        if predict == u.label :
            correct += 1

    #print("correct : %d"%(correct))
    return correct * (1/m)
    
    
    
```

```python
weight = [[-0.5],[-0.5],[0]]
loss_log_train = []
loss_log_test = []
iter_log = []

acc_log_train = []
acc_log_test = []


alpha =0.0001
m = train_pool_size
loss = 0

# for
i = 0
print(weight)

prev_loss = 0
cur_loss = 5


for i in range (0,8000) :
    h = hypothesis(weight,X)
    h_test = hypothesis(weight,X_test)
    
    gradient = np.dot(X.T, h-y) / y.size
    weight[:2] -= alpha*gradient    
    
    cur_train_loss = lossFunction(h,y)
    test_loss = lossFunction(h_test,y_test)
    acc = predict(weight,train_pool)
    acc_test = predict(weight, test_pool)
    
    loss_log_train.append(cur_train_loss)
    loss_log_test.append(test_loss)
    iter_log.append(i)
    acc_log_train.append(acc)
    acc_log_test.append(acc_test)
    
    print("loss: %f / %f , iter : %d, train_acc : %f / test_acc : %f"%(cur_train_loss,test_loss,i,acc,acc_test))
    
```

```python
t1 = plt.plot(iter_log,loss_log_train, color='orange',label='Training Loss')
t2 = plt.plot(iter_log,loss_log_test, color= 'red',label='Testing Loss')
plt.title("Loss")
plt.legend(['Training Loss','Testing Loss'])
plt.show()
```

```python
t1 = plt.plot(iter_log,acc_log_train, color='orange',label='Training Acc')
t2 = plt.plot(iter_log,acc_log_test, color= 'red',label='Testing Acc')
plt.title("Accuracy")
plt.legend(['Training Acc','Testing Acc'])
plt.show()
```
