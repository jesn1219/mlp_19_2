```python
''' Hi , '''
import matplotlib.pyplot as plt
import math
import numpy as np
import random
```

```python
print("Hello")
train_pool_size = 300
test_pool_size = 300



```

```python
class point :
    def __init__(self,x,y) :
        self.set_x(x)
        self.set_y(y)
        print("Point is setted. X : %d | Y : %d"%(self.x,self.y))

    def set_x(self,x) :
        self.x = x
    def set_y(self,y) :
        self.y = y
    def get_x(self) :
        return self.x
    def get_y(self) :
        return self.y
```

```python
class unit :
    def __init__(self, x, y, label) :
        self.x = x
        self.y = y
        self.label = label
        #print("test_unit is created. X : %d | Y : %d"%(self.x,self.y))
     
    def set_x(self,x) :
        self.x = x
    def set_y(self,y) :
        self.y = y
    def get_x(self) :
        return self.x
    def get_y(self) :
        return self.y    
        
```

```python
train_point_A = point(15,30)
train_point_B = point(30,15)

train_pool_a = []
train_pool_b = []
test_pool_a = []
test_pool_b = []
np.random.seed(100)

for i in range(0,train_pool_size) :
    random_x = train_point_A.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_A.get_y() + (5 * ((np.random.randn(1))[0]))
    train_pool_a.append(unit(random_x,random_y,0))
    random_x = train_point_B.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_B.get_y() + (5 * ((np.random.randn(1))[0]))
    train_pool_b.append(unit(random_x,random_y,1))
    
    
for i in range(0, test_pool_size) :
    random_x = train_point_A.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_A.get_y() + (5 * ((np.random.randn(1))[0]))
    test_pool_a.append(unit(random_x,random_y,0))
    random_x = train_point_B.get_x() + (5 * ((np.random.randn(1))[0]))
    random_y = train_point_B.get_y() + (5 * ((np.random.randn(1))[0]))
    test_pool_b.append(unit(random_x,random_y,1))
    
```

```python
train_pool = train_pool_a + train_pool_b
test_pool = test_pool_a + test_pool_b

x_train = [u.get_x() for u in train_pool]  
y_train = [u.get_y() for u in train_pool]  
x_test = [u.get_x() for u in test_pool]
y_test = [u.get_y() for u in test_pool]
```

```python
x_train_a = [u.get_x() for u in train_pool_a]
y_train_a = [u.get_y() for u in train_pool_a]
x_train_b = [u.get_x() for u in train_pool_b]
y_train_b = [u.get_y() for u in train_pool_b]

t1 = plt.scatter(x_train_a, y_train_a, color='blue', marker='o')
t2 = plt.scatter(x_train_b, y_train_b, color='black', marker='o')
plt.title("Two Clusters of Training dataSet")
plt.legend([t1,t2],['A','B'])
plt.show()


```

```python
x_test_a = [u.get_x() for u in test_pool_a]
y_test_a = [u.get_y() for u in test_pool_a]
x_test_b = [u.get_x() for u in test_pool_b]
y_test_b = [u.get_y() for u in test_pool_b]

t1 = plt.scatter(x_test_a, y_test_a, color='blue', marker='o')
t2 = plt.scatter(x_test_b, y_test_b, color='black', marker='o')
plt.title("Two Clusters of Testing dataSet")
plt.legend([t1,t2],['A','B'])
plt.show()


```

```python
t1 = plt.scatter(x_train,y_train, color='orange',marker='o')
t2 = plt.scatter(x_test, y_test, color= 'red', marker= 'o')
plt.title("Two Clusters of dataset")
plt.legend([t1,t2], ['Training dataset','testing dataset'])
plt.show()
```
